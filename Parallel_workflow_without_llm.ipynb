```python
from langgraph.graph import StateGraph, START,END
from typing import TypedDict
```

```python
class Cricket_calulator(TypedDict):
  Total_runs:int
  Total_balls:int
  Total_4s:int
  Total_6s:int
  strike_rate:float
  runs_in_boundary_percent:float
```

```python
def strike_rate(s:Cricket_calulator):
  Total_runs=s['Total_runs']
  Total_balls=s['Total_balls']
  s['strike_rate']=Total_runs/Total_balls*100
  return {'strike_rate':s['strike_rate']}
```

```python
def runs_in_boundary_percent(s:Cricket_calulator):
  Total_runs=s['Total_runs']
  Total_4s=s['Total_4s']
  Total_6s=s["Total_6s"]
  s['runs_in_boundary_percent']=(Total_4s*4+Total_6s*6)/Total_runs*100
  return {'runs_in_boundary_percent':s['runs_in_boundary_percent']}
```

```python
graph=StateGraph(Cricket_calulator)

graph.add_node("strike_rate",strike_rate)
graph.add_node("runs_in_boundary_percent",runs_in_boundary_percent)


graph.add_edge(START,"strike_rate")
graph.add_edge(START,"runs_in_boundary_percent")
graph.add_edge("strike_rate",END)
graph.add_edge("runs_in_boundary_percent",END)

workflow=graph.compile()
initial_state={'Total_runs':100,'Total_balls':50,'Total_4s':10, 'Total_6s':3}
final_state=workflow.invoke(initial_state)
print(final_state)
```

```python
with open("graph.png", "wb") as f:
    f.write(workflow.get_graph().draw_mermaid_png())
```
