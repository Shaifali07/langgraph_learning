```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict,Literal
```

```python
class RootNature(TypedDict):
  a:int
  b:int
  c:int
  delta:int
  root_nature:str
```

```python
def calculate_delta(RootNature):
  delta= RootNature["b"]**2-4*RootNature["a"]*RootNature["c"]
  return {"delta":delta}
def real_root(RootNature):
  root_nature="real_root"
  return {"root_nature":root_nature}
def repeated_root(RootNature):
  root_nature="repeated_root"
  return {"root_nature":root_nature}
def No_real_root(RootNature):
  root_nature="No_real_root"
  return {"root_nature":root_nature}
def checkcondition(RootNature)->Literal["real_root","repeated_root","No_real_root"]:
  if(RootNature["delta"]>0):
    return "real_root"
  elif(RootNature["delta"]==0):
    return "repeated_root"
  else:
    return "No_real_root"


```

```python
graph=StateGraph(RootNature)

graph.add_node("calculate_delta",calculate_delta)
graph.add_node("real_root",real_root)
graph.add_node("No_real_root",No_real_root)
graph.add_node("repeated_root",repeated_root)
graph.add_edge(START,"calculate_delta")
graph.add_conditional_edges("calculate_delta",checkcondition)
graph.add_edge("real_root",END)
graph.add_edge("No_real_root",END)
graph.add_edge("repeated_root",END)
workflow=graph.compile()
initial_state={"a":1,"b":2,"c":1}
final_state=workflow.invoke(initial_state)
print(final_state)
```
